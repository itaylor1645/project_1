# Gold & Silver Futures Analysis


```python
import matplotlib.pyplot as plt
import metalfutures as mf
```

```python
gold_futures_df = mf.get_gold_futures_df().set_index("date")
silver_futures_df = mf.get_silver_futures_df().set_index("date")
```

## Comparing Closing Price


```python
fig, (ax1, ax2) = plt.subplots(2, figsize=(8, 6))
fig.suptitle("Closing Price")
gold_futures_df["close"].plot.line(
    ax=ax1, title="Gold Futures", xlabel="Date", ylabel="Price", color="gold"
)
silver_futures_df["close"].plot.line(
    ax=ax2, title="Silver Futures", xlabel="Date", ylabel="Price", color="silver"
)
for ax in (ax1, ax2):
    ax.margins(x=0.005)
plt.tight_layout()
plt.show()
```

```python
print(
    "Correlation between gold and silver closing price:",
    gold_futures_df["close"].corr(silver_futures_df["close"]),
)
```

```python
fig, (ax1, ax2) = plt.subplots(2, figsize=(8, 6))
fig.suptitle("Daily Return of Closing Price")
gold_futures_df["close"].pct_change().dropna().plot.line(
    ax=ax1, title="Gold Futures", xlabel="Date", ylabel="Percent Return", color="gold"
)
silver_futures_df["close"].pct_change().dropna().plot.line(
    ax=ax2,
    title="Silver Futures",
    xlabel="Date",
    ylabel="Percent Return",
    color="silver",
)
for ax in (ax1, ax2):
    ax.margins(x=0.005)
    ax.grid(axis="y")
plt.tight_layout()
plt.show()
```

## Analyzing Patterns in Closing Price


```python
# TODO period analysis; weekly, yearly
fig, (ax1, ax2) = plt.subplots(2, figsize=(8, 6))
fig.suptitle("Average Closing Price per Weekday")
gold_futures_df["close"].groupby(
    gold_futures_df.index.isocalendar().day
).mean().plot.line(
    ax=ax1,
    title="Gold Futures",
    xlabel="Day of Week",
    ylabel="Average Price",
    color="gold",
)
silver_futures_df["close"].groupby(
    silver_futures_df.index.isocalendar().day
).mean().plot.line(
    ax=ax2,
    title="Silver Futures",
    xlabel="Day of Week",
    ylabel="Average Price",
    color="silver",
)
for ax in (ax1, ax2):
    ax.margins(x=0.005)
    ax.set_xticks(range(1, 6))
    ax.set_xticklabels(["Mon", "Tue", "Wed", "Thu", "Fri"])
plt.tight_layout()
plt.show()
```

```python
fig, (ax1, ax2) = plt.subplots(2, figsize=(8, 6))
fig.suptitle("Average Closing Price per Week of Year")
gold_futures_df["close"].groupby(
    gold_futures_df.index.isocalendar().week
).mean().plot.line(
    ax=ax1,
    title="Gold Futures",
    xlabel="Week of Year",
    ylabel="Average Price",
    color="gold",
)
silver_futures_df["close"].groupby(
    silver_futures_df.index.isocalendar().week
).mean().plot.line(
    ax=ax2,
    title="Silver Futures",
    xlabel="Week of Year",
    ylabel="Average Price",
    color="silver",
)
for ax in (ax1, ax2):
    ax.margins(x=0.005)
    ax.set_xticks(range(4, 54, 4))
plt.tight_layout()
plt.show()
```
